```python
# Load GBDT model and plot SHAP values
model = joblib.load(f"{file_path}\\final\\2d_combined_{target}_final_model.pkl")

features = joblib.load(f"{file_path}\\final\\2d_combined_{target}_final_features.pkl")

X_train = joblib.load(f"{file_path}\\df_train_{target}_engineered.pkl")
y_train = pd.DataFrame(X_train[target])
X_train = X_train[features['feature'].to_list()]

#check if SHAP file already exists
if os.path.exists(f'{output_file_path}//{target}_shap_values.pkl'):
    print("SHAP values file found. Loading SHAP values...")
    shap_values = joblib.load(f'{output_file_path}//{target}_shap_values.pkl')
else:
    print("Calculating SHAP values...")
    explainer = shap.Explainer(model)
    shap_values = explainer(X_train)
    joblib.dump(shap_values, f'{output_file_path}//{target}_shap_values.pkl')

# plot SHAP summary plot for most important features in the previous plot
shap.summary_plot(
    shap_values, 
    X_train, 
    show=False, 
    max_display=len(features), 
    plot_size=(12, 20)
    )

ax = plt.gca()

ax.tick_params(axis='x', labelsize=20)
ax.set_xlabel("SHAP value (impact on model output)", fontsize=20)


cbar = plt.gcf().axes[-1]
cbar.tick_params(labelsize=20)
cbar.set_ylabel(cbar.get_ylabel(), fontsize=20)

# --- step 2: get y positions and remove SHAP’s labels ---
yticks = ax.get_yticks()
ytick_strings = [label.get_text() for label in ax.get_yticklabels()]
ax.set_yticklabels([])     # hide original
ax.set_ylabel("")          # remove empty space

# --- step 3: draw custom labels with halo ---
for y, feature in zip(yticks, ytick_strings):
    txt = ax.text(
        -0.02, y, feature,
        ha='left', va='center',
        transform=ax.get_yaxis_transform(),
        fontsize=16, color='black',
        zorder=1000, alpha = 0.6        
    )

    # Apply halo
    txt.set_path_effects([
        PathEffects.Stroke(linewidth=3, foreground='white'),
        PathEffects.Normal()
    ])

# optionally tighten left margin
plt.subplots_adjust(left=0.22)

plt.tight_layout()

# Save the SHAP summary plot
plt.savefig(f'{output_file_path}//{target}_SHAP_all_features_plot.png', dpi=300)
plt.savefig(f'{output_file_path}//{target}_SHAP_all_features_plot.pdf', format='pdf')
```

```python
no_of_features_to_plot = 25

# mean absolute SHAP value across samples
mean_abs_shap = np.abs(shap_values.values).mean(axis=0)

# get indices of top N features
topN_idx = np.argsort(mean_abs_shap)[-no_of_features_to_plot:][::-1]

# build a new Explanation with ONLY those features
shap_values_topN = shap.Explanation(
    values=shap_values.values[:, topN_idx],
    base_values=shap_values.base_values,
    data=shap_values.data[:, topN_idx],
    feature_names=np.array(shap_values.feature_names)[topN_idx]
)

shap.summary_plot(
    shap_values_topN, 
    X_train[shap_values_topN.feature_names[:no_of_features_to_plot]], 
    show=False, 
    max_display=no_of_features_to_plot, 
    plot_size=(15, 12)
    )

ax = plt.gca()

ax.tick_params(axis='x', labelsize=20)
ax.set_xlabel("SHAP value (impact on model output)", fontsize=20)

cbar = plt.gcf().axes[-1]
cbar.tick_params(labelsize=20)
cbar.set_ylabel(cbar.get_ylabel(), fontsize=20)

# --- step 2: get y positions and remove SHAP’s labels ---
yticks = ax.get_yticks()
ytick_strings = [label.get_text() for label in ax.get_yticklabels()]
ax.set_yticklabels([])     # hide original
ax.set_ylabel("")          # remove empty space

# --- step 3: draw custom labels with halo ---
for y, feature in zip(yticks, ytick_strings):
    txt = ax.text(
        -0.02, y, feature,
        ha='left', va='center',
        transform=ax.get_yaxis_transform(),
        fontsize=14, color='black',
        zorder=1000, alpha = 0.6        
    )

    # Apply halo
    txt.set_path_effects([
        PathEffects.Stroke(linewidth=3, foreground='white'),
        PathEffects.Normal()
    ])

# optionally tighten left margin
plt.subplots_adjust(left=0.22)

plt.tight_layout()

# Save the SHAP summary plot
plt.savefig(f'{output_file_path}//{target}_SHAP_top_{no_of_features_to_plot}_features_from_SHAP_plot.png', dpi=300)
plt.savefig(f'{output_file_path}//{target}_SHAP_top_{no_of_features_to_plot}_features_from_SHAP_plot.pdf', format='pdf')
```
